```python
import healpy as h
```

```python
teb=h.read_map("../dataout/planck353_year1_TEB_map_nside512.fits",(0,1,2))
tebalm=h.map2alm(teb,lmax=1024,pol=False)
```

```python
for i in range(size(tebalm[0])):
    ell=h.Alm.getlm(1024,i)[0]
    if ell>1:
        f=sqrt(1./((ell+2)*(ell+1)*(ell-1)*ell))
    else:
        f=0.
    tebalm[1][i]=tebalm[1][i]*f
    tebalm[2][i]=tebalm[2][i]*f
    
teb=h.alm2map(tebalm,nside=512,pol=False)
```

```python
h.orthview(teb[1],rot=(0,90))
h.mollview(teb[1],rot=(0,0))
```

```python
mask=h.read_map("../datain/mask_gal70_nside512.fits",verbose=False)
badmask=zeros(size(mask),float) ; badmask[mask>1e-6]=1.
tqu=h.read_map("../datain/planck353_year1_TQU_map_nside512.fits",(0,1,2),verbose=False)
tqu[0][:]=h.remove_dipole(tqu[0][:])
tqu[1][:]=h.remove_monopole(tqu[1][:]*log10(11*badmask-1)) ; tqu[1][badmask==0]=0
tqu[2][:]=h.remove_monopole(tqu[2][:]*log10(11*badmask-1)) ; tqu[2][badmask==0]=0
teb1alm=h.map2alm(tqu*mask,lmax=1024,pol=True)
teb1=h.alm2map(teb1alm,nside=512,pol=False,verbose=False)
```

```python
h.orthview(teb1[1],rot=(0,90))
h.orthview(teb[1],rot=(0,90))
h.orthview(teb[1]-teb1[1],rot=(0,90))
```

```python
h.orthview(teb1[2],rot=(0,90))
h.orthview(teb[2],rot=(0,90))
h.orthview(teb[2]-teb1[2]*mask,rot=(0,90))
```

```python
h.mollview(teb[2]-teb1[2]*mask,rot=(0,90))
```

```python
cle=h.alm2cl(h.map2alm(teb[1],lmax=1024))
cle1=h.alm2cl(h.map2alm(teb1[1],lmax=1024))
```

```python
clb=h.alm2cl(h.map2alm(teb[2],lmax=1024))
clb1=h.alm2cl(h.map2alm(teb1[2],lmax=1024))
```

```python
ell=arange(1025)
f=ell*(ell+1)/(2.*pi)
plot(ell,f*cle,"r-")
plot(ell,f*cle1,"b-")

figure()
plot(ell,f*clb,"r-")
plot(ell,f*clb1,"b-")
```
