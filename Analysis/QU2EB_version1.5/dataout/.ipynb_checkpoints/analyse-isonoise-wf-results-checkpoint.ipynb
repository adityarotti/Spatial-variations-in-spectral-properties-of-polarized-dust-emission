```python
import healpy as h
```

```python
m=h.read_map("isonoise/wf_data_60.fits",verbose=False)
ncov=h.read_map("isonoise/ncov.fits",verbose=False)
d=h.read_map("../../datain/isonoise_data.fits",verbose=False)
s=h.read_map("../../datain/cmb.fits",verbose=False)
msk=h.read_map("../../datain/nside64_wmap_mask.fits",verbose=False)
```

```python
h.mollview(ncov)
h.mollview(ncov*msk)
```

```python
h.mollview(m,title="Weiner filtered map (WF)")
h.mollview(d*msk-s*msk,title="data - signal")
h.mollview(m-d*msk,title="WF map - data")
h.mollview(m-s*msk,title="WF map - signal")
```

```python
chi2=loadtxt("./isonoise/chi2.dat")
plot(chi2[:,0],chi2[:,1])
semilogx()
xlabel("Iteration")
ylabel("$\chi^2$")
```

```python
ell=arange(0,129,1)
plot(ell, ell*(ell+1)*h.alm2cl(h.map2alm(s,128))/(2.*pi),label="True Signal")
plot(ell,ell*(ell+1)*h.alm2cl(h.map2alm(d*msk,128))/(2.*pi),label="Data")
plot(ell, ell*(ell+1)*h.alm2cl(h.map2alm(s*msk,128))/(2.*pi),label="True Signal masked")
plot(ell,ell*(ell+1)*h.alm2cl(h.map2alm(d,128))/(2.*pi),label="Data unmasked")
plot(ell, ell*(ell+1)*h.alm2cl(h.map2alm(m,128))/(2.*pi),label="WF signal")

legend(loc=0)
```
