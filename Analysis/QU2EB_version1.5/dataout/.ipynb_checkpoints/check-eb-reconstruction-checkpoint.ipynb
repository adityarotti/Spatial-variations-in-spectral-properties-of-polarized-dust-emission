```python
import healpy as h
nside=512
ell=arange(2*nside+1)
```

```python
we=h.read_map("teb_masked.fits",1,verbose=False)
wb=h.read_map("teb_masked.fits",2,verbose=False)
apomask=h.read_map("../../../planck_data/dgraded_data/mask_gal70_nside512.fits",verbose=False)
e=h.read_map("teb_fullsky.fits",1,verbose=False)
b=h.read_map("teb_fullsky.fits",2,verbose=False)
```

```python
clwb=h.alm2cl(h.map2alm(wb,2*nside))/((ell+2)*(ell-1)*(2.*pi))
clfsb=h.alm2cl(h.map2alm(b,2*nside))/((ell+2)*(ell-1)*(2.*pi))
clbm=h.alm2cl(h.map2alm(b*apomask*apomask,2*nside))/((ell+2)*(ell-1)*(2.*pi))

cl=h.read_cl("../datain/r0p1/cl.fits")
clbbrp1=ell*(ell+1)*cl[2][:2*nside+1]/(2.*pi)
```

```python
plot(ell,clwb,label="Residual corrected B")
plot(ell,clbm,label="Fullsky B masked")
plot(ell,clfsb,label="Fullsky B")

plot(ell,clbbrp1,label="$r=0.1$")
#plot(ell,clbbrp1*1e-2,label="$r=10^{-3}$")
#plot(ell,clbbrp1*1e-4,label="$r=10^{-5}$")
#plot(ell,clbbrp1*1e-6,label="$r=10^{-7}$")
#plot(ell,clbbrp1*1e-8,label="$r=10^{-9}$")
#plot(ell,lensbb,"k-",label="Lensing")

xlabel("multipole, $\ell$")
ylabel("$\ell(\ell+1)C^{BB}_{\ell}/(2\pi)$")
loglog()
#xlim(0,1e6)
#ylim(1e-12,e-1)
legend(bbox_to_anchor=(1.55, 1.025))
legend(loc="upper right",fontsize=10)

savefig("b-reconstruction-with-mask.png", dpi=200)
```

```python
plot(ell,clwb,label="Residual corrected B")
plot(ell,clbm,label="Masked B")
plot(ell,clfsb,label="Fullsky B")
#plot(ell,clbbrp1*0.5,label="$r=0.05$")

xlabel("multipole, $\ell$")
ylabel("$\ell(\ell+1)C^{BB}_{\ell}/(2\pi)$")
loglog()
xlim(0,1100)
#ylim(1e-12,5e-8)
legend(bbox_to_anchor=(1.55, 1.025))
legend(loc=0,fontsize=10)

savefig("b-spectra-match.png", dpi=200)

figure()
lmax=2050
plot(ell[:lmax],clwb[:lmax]/clbm[:lmax])

ylabel("$C_{\ell}^{MB}/C_{\ell}^{BM}$")
xlabel("multipole, $\ell$")

#hlines(1,0,lmax)
ylim(0.995,1.01)
xlim(0,1000)
#semilogy()
savefig("b-spectra-ratio.png", dpi=200)
```

```python
clwe=h.alm2cl(h.map2alm(we,2*nside))/((ell+2)*(ell-1)*(2.*pi))
clem=h.alm2cl(h.map2alm(e*apomask*apomask,2*nside))/((ell+2)*(ell-1)*(2.*pi))
clfse=h.alm2cl(h.map2alm(e,2*nside))/((ell+2)*(ell-1)*(2.*pi))
```

```python
plot(ell,clwe,label="Residual corrected E")
plot(ell,clem,label="Masked E")
plot(ell,clfse,"-.",alpha=1.,label="Fullsky E masked")

loglog()
legend(bbox_to_anchor=(1.55, 1.025))
```
