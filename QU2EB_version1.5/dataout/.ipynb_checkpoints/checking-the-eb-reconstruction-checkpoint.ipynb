```python
import healpy as h
```

```python
nside=512
ell=arange(0,2*nside+1)
pwc=h.pixwin(nside)
pwc=pwc[:2*nside+1]
bl=h.gauss_beam((15.0/60.0)*(pi/180),lmax=2*nside)
bl[:]=1.0

# Reading in the theoretical angular power spectra.

cltt=h.read_cl("../datain/cl.fits")[0][:2*nside+1]
clee=h.read_cl("../datain/cl.fits")[1][:2*nside+1]
clbb=h.read_cl("../datain/cl.fits")[2][:2*nside+1]
clte=h.read_cl("../datain/cl.fits")[3][:2*nside+1]

data=loadtxt("../datain/clrp1.dat")
clbbrp1=append(zeros(2,float),data[:,3])
clbbrp1=clbbrp1[:2*nside+1]

# Reading in the masks
mask=h.read_map("../datain/wmap_polarization_analysis_mask_r9_7yr_v4.fits")
mask=h.read_map("mask.fits")
apomask=h.read_map("apomask.fits")
widemask=h.read_map("widemask.fits")
```

```python
filename="teb_fullsky.fits"
tmap=h.read_map(filename,0) ; ttfs=h.alm2cl(h.map2alm(tmap,lmax=2*nside))/(pwc*pwc*bl*bl)
emap=h.read_map(filename,1) ; eefs=h.alm2cl(h.map2alm(emap,lmax=2*nside))/(pwc*pwc*bl*bl)
bmap=h.read_map(filename,2) ; bbfs=h.alm2cl(h.map2alm(bmap,lmax=2*nside))/(pwc*pwc*bl*bl)
```

```python
plot(ell,ell*(ell+1)*ttfs/(2.*pi),alpha=0.4,label="SIM TT")
plot(ell,ell*(ell+1)*cltt/(2.*pi),alpha=1.0,label="THRY TT")

plot(ell,eefs/((ell+2)*(ell-1)*(2.*pi)),alpha=0.4,label="SIM EE")
plot(ell,ell*(ell+1)*clee/(2.*pi),alpha=1.0,label="THRY EE")

xlabel("multipole, $\ell$")
ylabel("$\ell(\ell+1)C_{\ell}$  $\mu K^2$")
loglog()
legend(bbox_to_anchor=(1.30, 1.025),title="Fullsky")
ylim(1e-3,1e4)

##########################################################

figure()
plot(ell,bbfs/((ell+2)*(ell-1)*(2.*pi)),alpha=0.4,label="SIM BB")
plot(ell,ell*(ell+1)*clbb/(2.*pi))

plot(ell,clbbrp1,label="$r=10^{-1}$")
plot(ell,clbbrp1*1e-3,label="$r=10^{-3}$")
plot(ell,clbbrp1*1e-6,label="$r=10^{-7}$")

xlabel("multipole, $\ell$")
ylabel("$\ell(\ell+1)C_{\ell}$  $\mu K^2$")
loglog()
legend(bbox_to_anchor=(1.28, 1.025),title="Fullsky")
```

```python
h.mollview(tmap)
h.mollview(emap)
h.mollview(bmap)
```

```python
filename="teb_masked.fits"
tmap_msk=h.read_map(filename,0)  ; tmap_msk=tmap_msk*widemask ; ttmsk=h.alm2cl(h.map2alm(tmap_msk,lmax=2*nside))/(pwc*pwc*bl*bl)
emap_msk=h.read_map(filename,1)  ; emap_msk=emap_msk*widemask ; eemsk=h.alm2cl(h.map2alm(emap_msk,lmax=2*nside))/(pwc*pwc*bl*bl)
bmap_msk=h.read_map(filename,2)  ; bmap_msk=bmap_msk*widemask ; bbmsk=h.alm2cl(h.map2alm(bmap_msk,lmax=2*nside))/(pwc*pwc*bl*bl)
```

```python
plot(ell,ell*(ell+1)*ttmsk/(2.*pi),alpha=0.4,label="SIM TT")
plot(ell,ell*(ell+1)*cltt/(2.*pi),alpha=1.0,label="THRY TT")

plot(ell,eemsk/((ell+2)*(ell-1)*(2.*pi)),alpha=0.4,label="SIM EE")
plot(ell,ell*(ell+1)*clee/(2.*pi),alpha=1.0,label="THRY EE")

xlabel("multipole, $\ell$")
ylabel("$\ell(\ell+1)C_{\ell}$  $\mu K^2$")
loglog()
legend(bbox_to_anchor=(1.30, 1.025),title="Masked")
ylim(1e-3,1e4)

##########################################################

figure()
plot(ell,bbmsk/((ell+2)*(ell-1)*(2.*pi)),alpha=0.4,label="SIM BB")
plot(ell,ell*(ell+1)*clbb/(2.*pi))

plot(ell,clbbrp1,label="$r=10^{-1}$")
plot(ell,clbbrp1*1e-3,label="$r=10^{-3}$")
plot(ell,clbbrp1*1e-6,label="$r=10^{-7}$")

loglog()
legend(bbox_to_anchor=(1.28, 1.025),title="Masked")
```

```python
h.mollview(tmap_msk)
h.mollview(emap_msk)
h.mollview(bmap_msk)
```

```python
filename="teb_apomasked.fits"
tmap_apomsk=h.read_map(filename,0)  ; tmap_apomsk=tmap_apomsk*widemask ; ttapomsk=h.alm2cl(h.map2alm(tmap_apomsk,lmax=2*nside))/(pwc*pwc*bl*bl)
emap_apomsk=h.read_map(filename,1)  ; emap_apomsk=emap_apomsk*widemask ; eeapomsk=h.alm2cl(h.map2alm(emap_apomsk,lmax=2*nside))/(pwc*pwc*bl*bl)
bmap_apomsk=h.read_map(filename,2)  ; bmap_apomsk=bmap_apomsk*widemask ; bbapomsk=h.alm2cl(h.map2alm(bmap_apomsk,lmax=2*nside))/(pwc*pwc*bl*bl)
```

```python
plot(ell,ell*(ell+1)*ttapomsk/(2.*pi),alpha=0.4,label="SIM TT")
plot(ell,ell*(ell+1)*cltt/(2.*pi),alpha=1.0,label="THRY TT")

plot(ell,eeapomsk/((ell+2)*(ell-1)*(2.*pi)),alpha=0.4,label="SIM EE")
plot(ell,ell*(ell+1)*clee/(2.*pi),alpha=1.0,label="THRY EE")

xlabel("multipole, $\ell$")
ylabel("$\ell(\ell+1)C_{\ell}$  $\mu K^2$")
loglog()
legend(bbox_to_anchor=(1.30, 1.025),title="Apo Masked")
ylim(1e-3,1e4)

##########################################################

figure()
plot(ell,bbapomsk/((ell+2)*(ell-1)*(2.*pi)),alpha=0.4,label="SIM BB")
plot(ell,ell*(ell+1)*clbb/(2.*pi))

plot(ell,clbbrp1,label="$r=10^{-1}$")
plot(ell,clbbrp1*1e-3,label="$r=10^{-3}$")
plot(ell,clbbrp1*1e-6,label="$r=10^{-7}$")

loglog()
legend(bbox_to_anchor=(1.28, 1.025),title="Apo Masked")
```
