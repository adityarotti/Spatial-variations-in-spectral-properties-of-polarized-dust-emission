```python
import healpy as h
nside=512
```

```python
mask=h.read_map("mask.fits")
apomask=h.read_map("apomask.fits")
widemask=h.read_map("widemask.fits")
```

```python
wmask=mask[:]*0+1.0
for i in range(size(wmask)):
    if abs(1.0-apomask[i])>=1e-6:
        wmask[i]=0.0
```

```python
der_apomask=h.alm2map_der1(h.map2alm(apomask,lmax=2*nside),nside,2*nside,2*nside)
der_mask=h.alm2map_der1(h.map2alm(mask,lmax=2*nside),nside,2*nside,2*nside)
```

```python
h.mollview(der_apomask[1]*wmask,title="Apodized")
h.mollview(der_mask[1]*wmask,title="Sharp")
```

```python
h.mollview(widemask-mask)
```
